```
from pyspark.sql.functions import *
from pyspark.sql.types import *
import glob, os ,sys
from collections import Counter
```

```
###Função para listar o diretório e o tipo de arquivo
def get_parquet_files(directory_path,tipo):
  arquivo_files = []
  files_to_treat = dbutils.fs.ls(directory_path)
  while files_to_treat:
    path = files_to_treat.pop(0).path
    if path.endswith('/'):
      files_to_treat += dbutils.fs.ls(path)
    elif path.endswith(f'.{tipo}'):
      arquivo_files.append(path)
  return arquivo_files
```

```
#criando a lista a ser lida
listagem = get_parquet_files("/mnt/stgdlmltg/bronze/LANDING_ZONE/SARA","parquet")
```

```
# filtrando regra de leitura da lista
lista_parquet = []
for l in listagem:
    if(l.count('/')) == 7:
        print(l)
        lista_parquet.append(l)        
```

```
#conferindo tamanho da lista total e lista a ser tratabalhada
print(len(listagem))
print(len(lista_parquet))
```

```
# splitando a lista a ser trabalhada
lista_splitada=[]
for l in listagem:
    if(l.count('/')) == 7:
        print(l.split('/'))
        lista_splitada.append(l.split('/'))
```

```
#len(lista_splitada)
#print(lista_splitada[0])
```

```
#Conferindo De:Para de leitura, gravação e tabela de dados no banco de dados na bronze
for l in lista_splitada:
    print(l[0]+'/'+l[1]+'/'+l[2]+'/'+l[3]+'/'+l[4]+'/'+l[5]+'/'+l[6]+'/'+l[7])#caminho leitura
    print(l[0]+'/'+l[1]+'/'+l[2]+'/'+l[3]+'/'+'RAW'+'/'+l[5]+'/'+l[6]+'/'+l[7].replace('.parquet','/'))#caminhoGravação
    print('bronze.'+l[5]+'_'+l[6]+'_'+l[7].replace('.parquet',''))#caminho_tabelas
```

```
item = 0
for l in lista_splitada:
    #lendo e contando registros da camanda de leitura
    fonte_parquet = (l[0]+'/'+l[1]+'/'+l[2]+'/'+l[3]+'/'+l[4]+'/'+l[5]+'/'+l[6]+'/'+l[7])
    print(f'Iniciando: indice:{item} do caminho:{fonte_parquet} \n')
    df = spark.read\
        .format("parquet")\
        .load(fonte_parquet)
    print (f'Leitura de {df.count()} registros ok, do caminho {fonte_parquet}')

    #gravando delta no estorage
    caminho_gravacao = (l[0]+'/'+l[1]+'/'+l[2]+'/'+l[3]+'/'+'RAW'+'/'+l[5]+'/'+l[6]+'/'+l[7].replace('.parquet','/'))
    df.write.format('delta')\
        .mode("overwrite")\
        .option("overwriteSchema", "true")\
        .save(caminho_gravacao)
    print (f'Gravado {df.count()} registros, no caminho {caminho_gravacao}')

    #lendo delta
    df = spark.read\
        .format("delta")\
        .load(caminho_gravacao)
    print('leitura e gravação ok')
    
    #Criando Tabela com formato delta
    caminh_tbl = ('bronze.'+l[5]+'_'+l[6]+'_'+l[7].replace('.parquet',''))
    df.write.mode("overwrite")\
        .format('delta')\
        .option('path', caminho_gravacao)\
        .saveAsTable(caminh_tbl)
    print(f'Criação OK da Tbl cujo comando de seleção é = select * from {caminh_tbl} \n\n')
    item=item+1
```
